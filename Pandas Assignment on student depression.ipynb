# Question 1: What percentage of students show signs of moderate to severe depression?.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


# Simulated dataset
data = {
    'Student_ID': range(1, 101),
    'Depression_Score': [random.randint(0, 27) for _ in range(100)]
}
df = pd.DataFrame(data)

# Categorize depression levels
def categorize(score):
    if score <= 4:
        return 'Minimal'
    elif score <= 9:
        return 'Mild'
    elif score <= 14:
        return 'Moderate'
    elif score <= 19:
        return 'Moderately Severe'
    else:
        return 'Severe'

df['Depression_Level'] = df['Depression_Score'].apply(categorize)

# Count and percentage of each level
level_counts = df['Depression_Level'].value_counts(normalize=True) * 100

# plot
plt.figure(figsize=(8, 6))
level_counts.plot(kind='bar',color='skyblue')
plt.title('Percentage of Students by Depression Level')
plt.ylabel('Percentage')
plt.xlabel('Depression Level')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

# Question 2: is there a correlation between academic performance and depression scores?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Simulat dataset
random.seed(42)
data = {
    'Student_ID': range(1,
101),
    'Depression_Score':
[random.randint(0, 27) for _ in
range(100)],
    'Academic_Performance':
[round(random.uniform(2.0,
4.0), 2) for _ in range(100)]
}

df = pd.DataFrame(data)

# Calculation correlation
correlation = df['Depression_Score'].corr(df['Academic_Performance'])

# Plot
plt.figure(figsize=(8, 6))
sns.regplot(x='Depression_Score', y='Academic_Performance', 
data=df,

scatter_kws={'alpha': 0.6},
line_kws={'color': 'red'})
plt.title('Correlation: {:.2f} between Depression Score and GPA'.format(correlation))
plt.xlabel('Depression Score')
plt.ylabel('Academic Performance (GPA)')
plt.grid(True, linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

# Question 3: Do students with low suport report higher depression levels?.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Simulate data
random.seed(42)
data = {
    'Student_ID': range(1,
101),
    'Family_Support':
[random.choice(['Low',
'Medium', 'High']) for _ in
range(100)],
    'Depression_Score':
[random.randint(0, 27) for _ in
range(100)]
}

df = pd.DataFrame(data)

# Set 1: View average depression by family support
avg_by_support = df.groupby('Family_Support')['Depression_Score'].mean()
print("Average Depression Score by Family Support:\n", avg_by_support)

# Step 2: Plot
plt.figure(figsize=(7, 5))
sns.boxplot(x='Family_Support', y='Depression_Score', data=df, palette='Set3')
plt.title('Depression Score by Level of Family Support')
plt.xlabel('Family Support')
plt.ylabel('Depression Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
